```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout, Bidirectional, Attention, BatchNormalization, Input
from tensorflow.keras import regularizers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from keras_tuner import RandomSearch
from tensorflow.keras.callbacks import EarlyStopping
from statsmodels.tsa.arima.model import ARIMA
from scipy.stats import chi2
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import pandas_datareader.data as web
import datetime
import shap
```

```python
# Define Asset Classes
asset_classes = {
    "Equities": [ "AAPL"]
    # "Bonds": ["IEF","^FVX"], # 7-10 Year Treasury Bond ETF, Treasury Yield 5 Years
    # "Commodities": ["GC=F", "CL=F"], # Gold, Oil
    # "Crypto": ["BTC-USD", "ETH-USD"],
    # "Real Estate": ["VNQ"]# REIT ETF
}

ticker = "AAPL"
macro_indicators = ["GDP", "CPIAUCSL", "UNRATE", "FEDFUNDS", "GS10", "INDPRO", "PPIACO", "RSXFS", "HOUST", "PSAVERT"]
```

```python
# Value at Risk (VaR) Calculation
def compute_var(returns, confidence=0.95):
    var = np.percentile(returns, (1 - confidence) * 100)
    return var

# Conditional VaR (Expected Shortfall)
def compute_cvar(returns, confidence=0.95):
    var = compute_var(returns, confidence)
    return returns[returns < var].mean()
```

```python
def kupiec_test(violations, num_obs, confidence_level=0.95):
    """Kupiec’s Proportion of Failures (POF) Test for VaR Validation.
    
    Arguments:
    - violations: Number of times actual loss exceeded predicted VaR
    - num_obs: Total number of observations
    - confidence_level: VaR confidence level (default: 95%)
    
    Returns:
    - p-value: Probability that the model's failure rate is statistically correct
    """
    if violations == 0 or violations == num_obs:
        return 1.0  # If no violations, assume model is too conservative, return high p-value

    expected_violations = num_obs * (1 - confidence_level)

    # Ensure numerical stability by preventing log(0) calculations
    violations = max(violations, 1e-6)  
    non_violations = max(num_obs - violations, 1e-6)

    L_ratio = -2 * ((expected_violations * np.log(violations / num_obs)) +
                    ((num_obs - expected_violations) * np.log(non_violations / num_obs)))

    p_value = 1 - chi2.cdf(L_ratio, df=1)
    
    return max(p_value, 1e-6)  # Ensure p-value is never exactly zero
```

```python
# Data Fetching and Preprocessing

def fetch_data(asset_list):
    data = {}
    for asset in asset_list:
        stock = yf.Ticker(asset)
        data[asset] = stock.history(period="max")[['Open', 'High', 'Low', 'Close', 'Volume']]
    return data

def fetch_macro_data(indicators):
    start = datetime.datetime(1992, 1, 1)
    end = datetime.datetime.now()
    return {ind: web.DataReader(ind, 'fred', start, end) for ind in indicators}
```

```python
# Data Imputation & Cleaning
def preprocess_data(df):
    df.fillna(method='ffill', inplace=True)
    df.fillna(method='bfill', inplace=True)
    df.dropna(inplace=True)
    return df
```

```python
# Feature Engineering

def create_features(df, asset_name):
    df[f'{asset_name}_LogReturn'] = np.log(df['Close'] / df['Close'].shift(1))
    df[f'{asset_name}_RollingVolatility'] = df[f'{asset_name}_LogReturn'].rolling(window=30).std()
    df[f'{asset_name}_MaxDrawdown'] = df['Close'] / df['Close'].rolling(window=30).max() - 1
    df['Year'] = df.index.year
    df['Month'] = df.index.month
    df['Week'] = df.index.isocalendar().week
    df['DayOfWeek'] = df.index.dayofweek

    df[f'{asset_name}_Momentum'] = df['Close'].diff(10)
    df[f'{asset_name}_RSI'] = 100 - (100 / (1 + df['Close'].pct_change().rolling(14).mean() / df['Close'].pct_change().rolling(14).std()))
    

    # if asset_name in asset_classes["Equities"]:
    #     df[f'{asset_name}_Momentum'] = df['Close'].diff(10)
    #     df[f'{asset_name}_RSI'] = 100 - (100 / (1 + df['Close'].pct_change().rolling(14).mean() / df['Close'].pct_change().rolling(14).std()))

    # if asset_name in asset_classes["Bonds"]:
    #     df[f'{asset_name}_RateChange'] = df['Close'].diff(5)

    # if asset_name in asset_classes["Commodities"]:
    #     df[f'{asset_name}_InflationSensitivity'] = df['Close'].rolling(30).corr(df['CPIAUCSL'])

    # if asset_name in asset_classes["Crypto"]:
    #     df[f'{asset_name}_Volatility_7d'] = df[f'{asset_name}_LogReturn'].rolling(7).std()
    #     df[f'{asset_name}_Sharpe_Ratio'] = df[f'{asset_name}_LogReturn'].mean() / df[f'{asset_name}_LogReturn'].std()

    return df
```

```python
def compute_shap_importance(model, X_sample):
    explainer = shap.Explainer(model, X_sample)
    shap_values = explainer(X_sample)

    # Plot feature importance
    shap.summary_plot(shap_values, X_sample)
```

```python
# Train-Test Split & Scaling
def train_test_split_data(full_df, asset_name):
    features = full_df.columns.difference([f'{asset_name}_LogReturn'])
    X = full_df[features].values
    y = full_df[f'{asset_name}_LogReturn'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    return X_train_scaled, X_test_scaled, y_train, y_test
```

```python
# Baseline Models (ARIMA & GARCH)
def arima_baseline(y_train, y_test):
    model = ARIMA(y_train, order=(5,1,0)).fit()
    return model.forecast(steps=len(y_test))

def linear_regression_baseline(X_train, X_test, y_train):
    model = LinearRegression()
    model.fit(X_train, y_train)
    return model.predict(X_test)
```

```python
# Modeling

def build_lstm_model(input_shape):
    print(f"inputshape: {input_shape}")
    inputs = Input(shape=(input_shape[0], input_shape[1]))
    
    # LSTM with Attention Mechanism
    lstm_out = Bidirectional(LSTM(64, return_sequences=True, dropout=0.3, recurrent_dropout=0.2))(inputs)
    
    # Attention Layer
    attention = Attention()([lstm_out, lstm_out])
    
    # More LSTM layers for better pattern extraction
    lstm_out = LSTM(32, return_sequences=False, dropout=0.2, recurrent_dropout=0.2)(attention)
    lstm_out = BatchNormalization()(lstm_out)
    
    # Fully Connected Layers
    dense_out = Dense(32, activation="relu")(lstm_out)
    dense_out = Dropout(0.3)(dense_out)
    dense_out = Dense(16, activation="relu")(dense_out)
    
    # Final Output Layer
    output = Dense(1)(dense_out)

    model = Model(inputs=inputs, outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="mse")
    
    return model
```

```python
# Performance Metrics & Backtesting
def evaluate_model(y_true, y_pred):
    print("Mean Absolute Error:", mean_absolute_error(y_true, y_pred))
    print("Mean Squared Error:", mean_squared_error(y_true, y_pred))
    print("R-Squared Score:", r2_score(y_true, y_pred))


def backtest_strategy(y_test, y_pred, asset_name):
    # Ensure y_pred has the same index as y_test
    if isinstance(y_pred, pd.Series):
        y_pred_series = y_pred  # No need to flatten if already a Series
    else:
        y_pred_series = pd.Series(y_pred.flatten(), index=y_test.index)  # Flatten only NumPy arrays
    plt.figure(figsize=(12,6))
    plt.plot(y_test.index, y_test, label='Actual', color='blue')
    plt.plot(y_pred_series.index, y_pred_series, label='Predicted', color='red', linestyle='dashed')
    plt.xlabel("Time")
    plt.ylabel("Log Return")
    plt.title(f"{asset_name} Model Backtest Performance")
    plt.legend()
    plt.show()
```

```python
# Final Processing for Each Asset Class

stock_data = fetch_data([ticker])[ticker]
macro_data = fetch_macro_data(macro_indicators)
full_df = preprocess_data(create_features(stock_data, ticker))
X_train, X_test, y_train, y_test = train_test_split(full_df.drop(columns=[f'{ticker}_LogReturn']), full_df[f'{ticker}_LogReturn'], test_size=0.2, shuffle=False)

input_shape = (X_train.shape[1], 1)
model = build_lstm_model(input_shape)
arima_pred = arima_baseline(y_train, y_test)
lr_pred = linear_regression_baseline(X_train, X_test, y_train)

X_train_reshaped = np.array(X_train, dtype=np.float32).reshape((X_train.shape[0], X_train.shape[1], 1))
X_test_reshaped = np.array(X_test, dtype=np.float32).reshape((X_test.shape[0], X_test.shape[1], 1))

history = model.fit(X_train_reshaped, np.array(y_train, dtype=np.float32),
                    epochs=50, batch_size=32, validation_split=0.2, 
                    callbacks=[EarlyStopping(monitor='val_loss', patience=5)])
y_pred = model.predict(X_test_reshaped)

print(f"Comparison for {ticker}:")
print("LSTM Model:")
evaluate_model(y_test, y_pred)
print("ARIMA Model:")
evaluate_model(y_test, arima_pred)
print("Linear Regression Model:")
evaluate_model(y_test, lr_pred)


# Compute VaR and Kupiec Test
var = compute_var(y_test)
violations = (y_test < var).sum()
kupiec_p_value = kupiec_test(violations, len(y_test))
cvar = compute_cvar(y_test)
print(f"{ticker} Value-at-Risk (VaR): {var}")
print(f"{ticker} Conditional VaR (CVaR): {cvar}")
print(f"{ticker} Kupiec Test p-value: {kupiec_p_value}")

evaluate_model(y_test, y_pred)
backtest_strategy(y_test, y_pred, ticker)
backtest_strategy(y_test, arima_pred, f"{ticker} - ARIMA")
backtest_strategy(y_test, lr_pred, f"{ticker} - Linear Regression")

# Plot Training & Validation Loss
plt.figure(figsize=(10,5))
plt.plot(history.history['loss'], label='Training Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='red', linestyle='dashed')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Model Training Loss Over Time")
plt.legend()
plt.show()
```
